Write a python program to display all the header tags from wikipedia.org and make data frame

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#sending request to the web page to get the source code
page  = requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
page
```

```python
#page content
soup = BeautifulSoup(page.content)
soup
```

```python
#scraping the headers from the page
headers = soup.find_all('h2',class_="mp-h2")
headers
```

```python
header_tags = []
```

```python
for i in headers:
    header_tags.append(i.text)
```

```python
header_tags
```

```python
h_tags = pd.DataFrame({"Headers":header_tags})
```

```python
h_tags
```

```python
news_link = []
nl = soup.find_all('a',class_="LatestNews-headline")
for i in nl:
    links = i.get("href")
    news_link.append(links)
```

```python
news =  pd.DataFrame({"Headline":headlines,"Time":time, "News Link":news_link})
news
```
